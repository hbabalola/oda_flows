```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading data
crs_data = pd.read_csv("crs_data.csv")
```

```python
#understanding variables
crs_data.head()
```

```python
crs_data.info()
```

```python
%matplotlib inline

#create a bar chart
plt.figure(figsize=(8,5)) 
plt.barh(crs_data['year'],crs_data['value'])

#Add titles and labels
plt.title("ODA disbursement by year")
plt.xlabel("Year")
plt.ylabel("Disbursement")

#show chart
plt.show()
```
